```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('/content/drive/MyDrive/koredepsac/psacquant.csv')
df
```

```python
biometrics = df.iloc[:, 6:8].values
egg_count = df.sheggcnt.values
egg_count

# splitting the dataset into the training and test set

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test  = train_test_split(biometrics, egg_count, test_size = 0.2, random_state = 0)

x_train
x_test
```

```python
# using decision tree to make sense out of the urological observations : proteinuria, hematuria, leucocyturia, glycosuria and nitrituria

haem_para = df[['haematuria','glycosuria','proteinuria','leukocyturia']].values # these are other morbidities associated with schistosomiasis
haem_para.reshape(-1,1)
u_schisto = df[['sheggcnt']].values
u_schisto

# importing the decision tree model
from sklearn.tree import DecisionTreeRegressor

dtr = DecisionTreeRegressor(random_state=0)
model = dtr.fit(haem_para, u_schisto)
# haem_para
pred_schist = dtr.predict([[3,3,3,0]]) # note that the double angular brakets are required for the four predictors to predict new schisto values

# visualizing the the decision tree regression results( higher resolution)
from sklearn import tree
import graphviz
dot_data = tree.export_graphviz(model, out_file='tree.dot')
dot_data

# eig_val, eig_vec = np.linalg.eig([[haem_para]])
```

```python
# using decision tree to make sense out of the urological observations : proteinuria, hematuria, leucocyturia, glycosuria and nitrituria

haem_para = df[['haematuria','glycosuria','proteinuria','leukocyturia']].values # these are other morbidities associated with schistosomiasis
haem_para.reshape(-1,1)
u_schisto = df[['sheggcnt']].values
u_schisto

from sklearn.model_selection import train_test_split


# importing the decision tree model
from sklearn.tree import DecisionTreeRegressor

dtr = DecisionTreeRegressor(max_depth=4, min_impurity_decrease=5, max_leaf_nodes=10)
model = dtr.fit(haem_para, u_schisto)
# haem_para
pred_schist = dtr.predict([[3,3,3,0]]) # note that the double angular brakets are required for the four predictors to predict new schisto values

# visualizing the the decision tree regression results( higher resolution)
from sklearn import tree
import graphviz
import pydot
import pydotplus
dot_data = tree.export_graphviz(model, out_file=None)
dot_data

graph = pydotplus.graph_from_dot_data(dot_data)

graph.write_pdf("psac_korede.pdf")
```

```python
# This is the multiple linear regression still observing the relationship between these four predictors with schistosomiasis
# N.B that the schistosoma eggcount used here is not normalized
haem_para = df[['haematuria','glycosuria','proteinuria','leukocyturia']].values # these are other morbidities associated with schistosomiasis
haem_para.reshape(-1,1)
u_schisto = df[['sheggcnt']].values
u_schisto

x_train,x_test, y_train, y_test = train_test_split(haem_para, u_schisto, test_size=0.2, random_state=42)

from sklearn.linear_model import LinearRegression
lr = LinearRegression()

lr_model = lr.fit(x_train, y_train)

prediction = lr.predict(x_test)
prediction
# plt.scatter(x_test,y_test, color="red")

# Now to know the model accuracy in order to ascertain if its a good model

from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, prediction)
# computing the regression coeficient, mean squared error and the variance score.. N.B variance score 1 is perfect

print(lr.coef_)
print(mse)
print(r2_score(y_test, prediction))
```
